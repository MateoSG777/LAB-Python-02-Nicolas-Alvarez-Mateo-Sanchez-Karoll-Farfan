## $\color{red}{\text{Ejercicio 6 - Minimax}}$


Dado el juego conocido como NIM, implementar una solución a este usando el algoritmo minimax visto en clase

En este juego hay una pila de piedras (por ejemplo 10).

En cada turno un jugador puede quitar 1, 2 o 3 piedras.

Pierde el jugador que no puede mover (porque ya no quedan piedras).

MAX = Computadora

MIN = Humano

Dado lo anterior y el algoritmo visto en clase, modelar el juego y simularlo donde MAX sera la cpu y MIN el humado el cual interactua mediante el input, el primer turno sera el de MIN

```

def print_state(piedras):
    print(f"\nPiedras restantes: {piedras}\n")

def legal_moves(piedras):
    return [m for m in [1, 2, 3] if m <= piedras]

def apply_move(piedras, mov):
    return piedras - mov


def terminal(piedras):
    return piedras == 0


def utility(piedras, maximizing):

    if terminal(piedras):
        return -1 if maximizing else 1
    return 0

def minimax(piedras, maximizing):

    if terminal(piedras):
        return utility(piedras, maximizing)

    if maximizing:
        best_value = float('-inf')
        for mov in legal_moves(piedras):
            nuevo_estado = apply_move(piedras, mov)
            value = minimax(nuevo_estado, False)
            best_value = max(best_value, value)
        return best_value


    else:
        best_value = float('inf')
        for mov in legal_moves(piedras):
            nuevo_estado = apply_move(piedras, mov)
            value = minimax(nuevo_estado, True)
            best_value = min(best_value, value)
        return best_value

def best_move(piedras):
    best_value = float('-inf')
    move_chosen = None

    for mov in legal_moves(piedras):
        value = minimax(apply_move(piedras, mov), False)
        if value > best_value:
            best_value = value
            move_chosen = mov

    return move_chosen


if __name__ == "__main__":

    piedras = 10
    turno_max = False

    print("=== JUEGO DE NIM ===")
    print("Hay 10 piedras. Puedes quitar 1, 2 o 3.")
    print("Pierde quien no puede mover.\n")

    while not terminal(piedras):
        print_state(piedras)


        if not turno_max:
            mov = int(input("Tu movimiento (1-3): "))
            while mov not in legal_moves(piedras):
                print("Movimiento inválido.")
                mov = int(input("Ingresa 1, 2 o 3: "))
            piedras = apply_move(piedras, mov)

        else:

            mov = best_move(piedras)
            print(f"La CPU quita {mov} piedras.")
            piedras = apply_move(piedras, mov)

        turno_max = not turno_max


    print_state(piedras)
    if turno_max:
        print("GANASTE (la CPU no pudo mover)")
    else:
        print("PERDISTE (no puedes mover)")
```
